```python
import gradio as gr
from langchain.document_loaders import PyPDFLoader
from langchain.embeddings import HuggingFaceEmbeddings
from langchain_google_genai import ChatGoogleGenerativeAI
from langchain import FAISS
```

```python
def process_pdf(pdf_path):
    loader = PyPDFLoader(pdf_path)
    pages = loader.load_and_split()

    embeddings = HuggingFaceEmbeddings(model_name="all-MiniLM-L6-v2")
    db = FAISS.from_documents(pages, embeddings)

    return db

db = process_pdf("/content/policy-booklet-0923 (1).pdf")

def answer_question(query):
    docs = db.similarity_search(query)
    content = "\n".join([x.page_content for x in docs])
    qa_prompt = "Use the following pieces of context to answer the user's question. If you don't know the answer, just say that you don't know, don't try to make up an answer.----------------"
    input_text = qa_prompt + "\nContext:" + content + "\nUser question:\n" + query
    llm = ChatGoogleGenerativeAI(model="gemini-pro")
    result = llm.invoke(input_text)
    return result.content

def upload_pdf(pdf):
    global db
    db = process_pdf(pdf.name)
    return "PDF uploaded and processed successfully. You can now ask questions."

with gr.Blocks() as demo:
    gr.Markdown("# PDF Question Answering System")
    pdf_input = gr.File(label="Upload a PDF")
    upload_button = gr.Button("Upload PDF")
    upload_button.click(upload_pdf, inputs=pdf_input, outputs=gr.Textbox())

    query_input = gr.Textbox(label="Ask a question based on the PDF content")
    answer_output = gr.Textbox(label="Answer")
    query_button = gr.Button("Ask")
    query_button.click(answer_question, inputs=query_input, outputs=answer_output)

demo.launch()
```
